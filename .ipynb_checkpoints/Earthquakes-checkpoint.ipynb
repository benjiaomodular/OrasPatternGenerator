```python
from datetime import date, timedelta
import pandas as pd
```

## Read earthquakes DB

```python
earthquakes_df = pd.read_csv('data/earthquakes.csv').sort_values('date')
earthquakes_df
```

```python
list(earthquakes_df)
```

```python
earthquakes_filtered_df = earthquakes_df.dropna(subset=['year', 'month', 'day'])[['year', 'month', 'day', 'intensity', 'magnitude']]
earthquakes_filtered_df = earthquakes_filtered_df[earthquakes_filtered_df['year'] > 1900]

earthquakes_filtered_df['date'] = earthquakes_filtered_df[["year", "month", "day"]].apply(
    lambda x: '-'.join(x.values.astype(int).astype(str)), axis="columns")

earthquakes_filtered_df['date']=pd.to_datetime(earthquakes_filtered_df['date'])
earthquakes_filtered_df[:10]
```

## Iterate daily from start date

```python
start_date = date(1970, 1, 1)
end_date = date(2020, 1, 1)
delta = timedelta(days=1)

earthquake_flags = []
while start_date <= end_date:
    start_date += delta
    
    has_earthquake = earthquakes_filtered_df.loc[earthquakes_filtered_df['date'] == start_date.strftime("%Y-%m-%d")]
    earthquake_flags.append(len(has_earthquake))
```

```python
with open("data/earthquakes.txt", 'a') as f:
    for earthquake_flag in earthquake_flags:
        f.write("\n{ef}{ef}{ef}".format(ef=earthquake_flag))
```
